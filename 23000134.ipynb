```python
data = {
    'Year': [2021, 2021, 2021, 2021,2021, 2021,2021, 2021, 2022, 2022, 2022, 2022, 2022, 2022,2022, 2022,],
    'Quarter': ['Q1', 'Q1', 'Q2', 'Q2','Q3', 'Q3','Q4', 'Q4', 'Q1', 'Q1', 'Q2', 'Q2', 'Q3', 'Q3','Q4', 'Q4'],
    'Metric': ['Sales', 'Expenses', 'Sales', 'Expenses', 'Sales', 'Expenses', 'Sales', 'Expenses', 'Sales', 'Expenses', 'Sales', 'Expenses','Sales', 'Expenses','Sales', 'Expenses'],
    'Channel': ['Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online', 'Online'],
    'Value': [10000, 5000, 12000, 6000, 15000, 7000, 18000, 8000,30000,42000,12000,91000,40000,60000,90000,80000]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df = pd.concat([df, df.copy().assign(Channel='In-Store', Value=[8000, 4000, 10000, 5000, 12000, 6000, 15000, 7000, 12000, 6000, 15000, 7000, 15000, 7000, 18000, 8000])])
```

```python
df
```

```python
df = df.pivot_table(index=['Year', 'Quarter', 'Channel'], columns='Metric', values='Value')

```

```python
df
```

```python
print("Online sales for Q1 2022:", df.loc[(2022, 'Q1', 'Online'), 'Sales'])
```

```python
print("In-store expenses in Q3 2021:", df.loc[(2021, 'Q3', 'In-Store'), 'Expenses'])
```

```python
print("Total sales for Q4 2021:", df.loc[(2021, 'Q4'), 'Sales'].sum())
```

```python
print("Expenses for the online channel in 2022:", df.loc[(2022, slice(None), 'Online'), 'Expenses'].sum())
```

```python
print("Sales data for Q2 2021:\n", df.loc[(2021, 'Q2'), 'Sales'])
```

```python
print("Maximum expense recorded in 2022:", df.loc[(2022), 'Expenses'].max())
```
